It is evident that the month of the year plans a big role in the burned acreage; Hence, it is importnat to convert the month to a dummy variable 

```python
# create job dummies(will add it to the result  dataeframe later )
result_dummies=pd.get_dummies(result.month, prefix='month')
# print 5 random rows
result_dummies.sample(n=5, random_state=1)
```

However, we only need 11 dummy variables (not 12), and thus we'll drop the first dummy variable.


```python
# drop the first column
result_dummies.drop(result_dummies.columns[0], axis=1, inplace=True)
result_dummies.sample(n=5, random_state=1)
```

```python
# concatenate the original DataFrame and the dummy DataFrame (axis=0 means rows, axis=1 means columns)
result = pd.concat([result, result_dummies], axis=1) # VERY IMPORTANT becasue otehrwise the data is added NaN at the bottm so use Axis=1 

# print 5 random rows
result.sample(n=5, random_state=1)
```
